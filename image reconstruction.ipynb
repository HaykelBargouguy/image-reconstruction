```python
def plot_MNIST_sample(X):
  """
  Plots 9 images in the MNIST dataset.

  Args:
     X (numpy array of floats) : Data matrix each column corresponds to a
                                 different random variable
  Returns:
    Nothing.

  """
  fig, ax = plt.subplots()
  k = 0
  for k1 in range(3):
    for k2 in range(3):
      k = k + 1
      plt.imshow(np.reshape(X[k, :], (28, 28)),
                 extent=[(k1 + 1) * 28, k1 * 28, (k2+1) * 28, k2 * 28],
                 vmin=0, vmax=255)
  plt.xlim((3 * 28, 0))
  plt.ylim((3 * 28, 0))
  plt.tick_params(axis='both', which='both', bottom=False, top=False,
                  labelbottom=False)
  plt.clim([0, 250])
  ax.set_xticks([])
  ax.set_yticks([])
  plt.show()
```

## Question
Dans la cellule ci dessous, j'ai visualisé scaler.scale_ et j'ai introduit scaler.scale_ .A quoi correspondent les éléments égaux à 1 dans le vecteur scaler.scale_? 


# Réponse :
Les éléments égaux à 1 dans le vecteur scaler.scale_ correspondent aux colonnes de la matrice de données qui ont une variance égale à 1 avant la mise à l'échelle. Cela signifie que ces colonnes ne subissent aucune transformation pendant la mise à l'échelle, car leur variance est déjà à l'échelle. En d'autres termes, ces colonnes ont déjà une dispersion relative similaire aux autres colonnes, donc elles n'ont pas besoin d'être mises à l'échelle pour que les données soient comparables.


```python
from sklearn import preprocessing
scaler= preprocessing.StandardScaler().fit(Pixels)

#fit(Pixels) permet de calculer la variance , la moyenne et l'ecart type de chaque colonne et 
#les store dans un vecteur .



print('le vecteur de moyenne de chaque colonne :')
print(scaler.mean_)
print("*************************************************************************")
print('le vecteur de variance de chaque colonne : ')
print(scaler.var_)
print("*************************************************************************")
print('le vecteur de l\'écart type de chque colonne :')
print(scaler.scale_)

Pixels_scaled = scaler.transform(Pixels)
#-----> centrée-réduite
#transform(Pixels) : applique la transformation de mise à l'échelle centrée-réduite aux données
#d'entrée en soustrayant la moyenne de chaque colonne et en divisant chaque 
#colonne par son écart type

Pixels_scaled.mean(axis=0)
#Cette ligne de code peut être utile pour vérifier si la mise à l'échelle centrée-réduite a fonctionné
#correctement, car elle permet de s'assurer que la moyenne de chaque colonne est proche de zéro.
print(Pixels_scaled.mean(axis=0))
print("***********************************")




```

```python
#from sklearn.preprocessing import scale
#pca.fit(scale(Pixels))  (à tester....pour voir la différence avec StandardScaler()  !!!)



from sklearn.decomposition import PCA
PCA(copy=True, iterated_power='auto', n_components=None, random_state=None,
   svd_solver='auto', tol=0.0, whiten=False)
pca=PCA()
pca.fit(Pixels_scaled)

#les valeurs propres sont calculés à partie de la matrice de covariance des donnes centre-reduites(Pixels_scaled)

#Les valeurs propres sont stockées dans l'attribut explained_variance_ 

#Les valeurs propres sont classées en ordre décroissant par défaut par Scikit-learn.

#La variance expliquée par chaque composante principale est proportionnelle à la valeur propre correspondante

#Les valeurs propres sont une mesure de la quantité de variance totale des données qui est expliquée par chaque composante principale

#La variance expliquée par la i-ème axe ( dimension dans notre cas ) = λi / Σλi où λi est la i-ème valeur propre (eigenvalue dans notre cas)
```

```python
# Output de l'ACP, variance expliquée par axe, variance expliquée en %,
# variance cumulée en % pour les 'p' axes de vecteurs directeurs les vecteurs propres 
eig = pd.DataFrame(
    {   "Dimension" : ["Dim" + str(x + 1) for x in range(784)], 
        "Varianceexpliquée" : pca.explained_variance_,
     
     #La variance expliquée par chaque composante principale est une mesure de 
     #la quantité d'information contenue dans cette composante principale.
     
     #Plus une composante principale explique une grande partie de la variance totale des données,
     #plus elle est importante pour décrire les données.
     
     #En général, on retient les composantes principales qui expliquent une grande partie de la 
     #variance totale des données, tout en éliminant les composantes principales qui ne sont pas
     #importantes-----> Cela permet de réduire la dimensionnalité des données tout en conservant
     #l'information la plus importante pour la description des données.
     

        "%variance_expliquée" : np.round(pca.explained_variance_ratio_ * 100),
        "cumvarexpliquée" : np.round(np.cumsum(pca.explained_variance_ratio_) * 100)
     #La variable "cumvarexpliquée"  correspond à la variance cumulée expliquée par les premiers p axes de vecteurs propres .
     
     #Cette variable est calculée en prenant la somme cumulative des pourcentages de variance expliquée par les premiers p axes.
    }
)

eig
```

```python
def plot_variance_explained(variance_explained):
  """
  Plots eigenvalues.

  Args:
    variance_explained (numpy array of floats) : Vector of variance explained
                                                 for each PC
  Returns:
    Nothing.

  """
  plt.figure()
  plt.plot(np.arange(1, len(variance_explained) + 1), variance_explained,
           '--k')
  plt.xlabel('Number of components')
  plt.ylabel('Variance explained')
  plt.show()
```

```python
plot_variance_explained(pca.explained_variance_)

#Le graphe ci-dessous  montre une courbe décroissante qui commence généralement 
#par une forte variance expliquée pour les premières composantes principales,
#puis diminue à mesure que l'on ajoute plus de composantes principales.

    
    
```

```python
def plot_eigenvalues(evals, limit=True):
  """
  Plots eigenvalues.

  Args:
     (numpy array of floats) : Vector of eigenvalues

  Returns:
    Nothing.

  """

  plt.figure()
  plt.plot(np.arange(1, len(evals) + 1), evals, 'o-k')
  plt.xlabel('Component')
  plt.ylabel('Eigenvalue')
  plt.title('Scree plot')
  if limit:
    plt.show()
```

## Question
Proposer une correction de la procédure de reconstruction.

### Le probleme est au niveau de cette ligne :
Pixels_r=Pixels_r Pixels_r+np.dot(scaler.mean_.T.reshape(1,-1),np.ones(784).reshape(-1,1))

Dans ce cas , il s'agit d'un produit scalire entre deux vecteurs U*V de dimension respective (1,784) , (784,1)
=> On obtient alors un scaalire .

### Pour le corriger , 
On doit ajouter l'opérateur ( inverse ) sur les deux matrices ou bien on fait inverser ses ordres (V*U) comme indique la code suiavnt .

```python
#Le probleme est au niveau de cette ligne de code Pixels_r=Pixels_r +Pixels_r+np.dot(scaler.mean_.T.reshape(1,-1),np.ones(784).reshape(-1,1))
#dans ce cas , nous sommes en train de sommer les corfficients --> On obtient un scaalire .
#Pour corriger ce probleme on fait ,tout simplememnt , inverser l'ordre de ces deux matrices 

S_racine = np.diagflat( std_vector)

print(S_racine.shape)

print(scaler.mean_.reshape(1,-1).shape)

print(np.ones(784).reshape(-1,1).shape)

#on corrige la ligne suivante  ,
#Psi : contient les coordonnées des individus dans les axes de projections  calculer par PCA . 
#U : contient les vecteurs propres de la matrice de covariance des données centrées et réduites .

Pixels_r = np.dot(Psi[:, :K], U[:K,:])
print(Pixels_r)

#Ce produit scalaire permet de projeter les individus sur les K premiers axes 
#en ne gardant que les K premières composantes principales.
#la matrice S_racine correspond à la racine carrée de la matrice diagonale contenant les valeurs propres de la matrice de covariance
# et qui permet (S_racine) de "réduire" l'impact de chaque axe en fonction de sa variance expliquée ( car la racine carree les modère)

Pixels_r = np.dot(Pixels_r, S_racine)

#---> obtient les coordonnées des individus dans l'espace des K premiers axes , après avoir réduit l'impact de chaque axe
#en fonction de sa variance expliquée.
#on ajoute puis la moyenne de chaque variable (qui avait été soustraite lors de la mise à l'échelle initiale).

Pixels_r = Pixels_r + np.dot(np.ones((len(Pixels_r), 1)), scaler.mean_.reshape(1, -1))

Pixels_r.shape
```

```python
def plot_MNIST_reconstruction(X, X_reconstructed):
  """
  Plots 9 images in the MNIST dataset side-by-side with the reconstructed
  images.

  Args:
    X (numpy array of floats)               : Data matrix each column
                                              corresponds to a different
                                              random variable
    X_reconstructed (numpy array of floats) : Data matrix each column
                                              corresponds to a different
                                              random variable

  Returns: Nothing.
  """

  plt.figure()
  ax = plt.subplot(121)
  k = 0
  for k1 in range(3):
    for k2 in range(3):
      k = k + 1
      plt.imshow(np.reshape(X[k, :], (28, 28)),
                 extent=[(k1 + 1) * 28, k1 * 28, (k2 + 1) * 28, k2 * 28],
                 vmin=0, vmax=255)
  plt.xlim((3 * 28, 0))
  plt.ylim((3 * 28, 0))
  plt.tick_params(axis='both', which='both', bottom=False, top=False,
                  labelbottom=False)
  ax.set_xticks([])
  ax.set_yticks([])
  plt.title('Data')
  plt.clim([0, 250])
  ax = plt.subplot(122)
  k = 0
  for k1 in range(3):
    for k2 in range(3):
      k = k + 1
      plt.imshow(np.reshape(np.real(X_reconstructed[k, :]), (28, 28)),
                 extent=[(k1 + 1) * 28, k1 * 28, (k2 + 1) * 28, k2 * 28],
                 vmin=0, vmax=255)
  plt.xlim((3 * 28, 0))
  plt.ylim((3 * 28, 0))
  plt.tick_params(axis='both', which='both', bottom=False, top=False,
                  labelbottom=False)
  ax.set_xticks([])
  ax.set_yticks([])
  plt.clim([0, 250])
  plt.title('Reconstructed')
  plt.tight_layout()
```

Calculer les corrélation entre les variables et les axes factoriel. 
Dessiner le cercle decorrélation et visualiser la variables les plus corrélés 
avec les deux premiers axes DIm1  et Dim2

```python
plt.tight_layout()
eigval = (n - 1) / n * pca.explained_variance_
sqrt_eigval = np.sqrt(eigval)
corvar = np.zeros((p, p))

for k in range(p):
    corvar[:, k] = pca.components_[k, :] * sqrt_eigval[k]
corvar.shape


coordvar = pd.DataFrame({"id": X.drop(columns=["class"]).columns, "COR_1": corvar[:, 0], "COR_2": corvar[:, 1]})
fig, axes = plt.subplots(figsize=(25, 25))
fig.suptitle("cercle des corrélations")
axes.set_xlim(-1, 1)
axes.set_ylim(-1, 1)
axes.axvline(x=0, color="lightgray", linestyle="-", linewidth=3)
axes.axhline(y=0, color="lightgray", linestyle="-", linewidth=3)
for j in range(784):
    axes.text(coordvar["COR_1"][j], coordvar["COR_2"][j], coordvar["id"][j], fontsize=12)
plt.gca().add_artist(plt.Circle((0, 0), 1, color="blue", fill=False))


plt.show()
```
